#### Subtask 2.2.1: Segmented Univariate Analysis with `Categorical Variables`  
Categorical variables are best to study while doing any kind of analysis to `segregate behaviours into different buckets`.
- `For Example:` There is a column __`Loan grade`__. A segmented univariate plot showing `loan grade` w.r.t `loan status` will give us some insights into the most frequently `Charged off` loan grades. 
- These insights will help determine whether to sanction/not sanction loan.

Here, We work on deriving insights from `segmented univariate analysis on the following categorical variables`. (Data definitions are given for quick understanding)
1. `grade:` LC assigned loan grade
2. `term`: The number of payments on the loan. Values are in months and can be either 36 or 60
3. `home_ownership`: The home ownership status provided by the borrower during registration
4. `purpose`: A category provided by the borrower for the loan request
5. `verification_status`: Indicates if income was verified
6. `addr_state`: The state praovided by the borrower in the loan application

8. `pub_rec`: Number of derogatory public records

__*NOTE:*__  <br>
For every categorical variable, we will compute a metric table with
1. `Percentage of fully paid loans` in particular category (Ex: (paid loans in grade A/total loans given in grade A)*100) 
2. `Percentage of charged off loans` in particular category (Ex:  (Charged off loans in grade A/total loans given in grade A)*100) 
    - These 2 values compliment each other and have a total of 100
3. Ratio of __`Charged Off` to `Fully Paid`__ *__The higher the ratio, the higher the risk__*


#### Part-1 : Graph-1 -  "GRADE" w.r.t "LOAN STATUS"
##### Grade:  Performance metrics

```python
# Percentage of paid and Charged off loans in a particular grade:
# 1.1 for status: "fully paid" and "charfed off"
grade_count_p = loan_past['grade'].value_counts()
df_grade_percent_p = round((grade_count_p/ loan_past.shape[0])*100,2)

# 1.2 for status: "fully paid"
grade_count_f = loan_fullypaid['grade'].value_counts()
df_grade_percent_f = round((grade_count_f/ grade_count_p)*100,2)

# 1.3 for status: "fully paid" and "charfed off"
grade_count_c = loan_chargeof['grade'].value_counts()
df_grade_percent_c = round((grade_count_c/ grade_count_p)*100,2)

# Ratio of Charged Off loans to Fully Paid 
df_grade_ratio = round((grade_count_c/ grade_count_f),2)

df_grade_percent = pd.DataFrame([grade_count_p,df_grade_percent_p,df_grade_percent_f,df_grade_percent_c,df_grade_ratio], 
                                index = ['Total loans given in each grade','Percent loan given in each grade','Percent Fully Paid','Percent Charged Off', 'Ratio of Charged Off to Paid'])


df_grade_percent1 = df_grade_percent.T
df_grade_percent1
```

#### Subtask 2.3.1: Deriving Metrics (Converting Quantitative Variables to Categories)

Quantitative variables are `continuos variables` and we cannot analyse w.r.t categorical variables like `loan status`.
Lets `convert some of the Numeric Categorical variables into Categorical Variables` for the detail analysis
We will create 3 metrics based on: __(The ranges are taken from the spread of percentiles from descriptive statistics)__
1. `Income Classification`
2. `Interest Rate Classification`
3. `Loan Amount Classificaiton`

We add these features to all 3 dataframes that we are using for analysis: `loan_past`, `loan_fullypaid`  and `loan_chargeof`.
Our aim is to use these features largely for bivariate analysis, but we will quickly do the univariate to see if we find some insights.

##### Feature on "ANNUAL INCOME": income_classification 

```python
def income_classification(annual_inc):
    if annual_inc >= 70000:
        return 'High'
    elif annual_inc >= 35000 and annual_inc < 70000:
        return 'Medium'
    elif annual_inc >= 10000 and annual_inc < 35000:
        return 'Low'
    elif annual_inc >= 1 and annual_inc < 10000:
        return 'Poor'
    else:
        return 'Not Known'

# Additional column is added in all 3 dataframes

loan_past['annual_inc_zone'] = loan_past['annual_inc'].apply(income_classification)
loan_fullypaid['annual_inc_zone'] = loan_fullypaid['annual_inc'].apply(income_classification)
loan_chargeof['annual_inc_zone'] = loan_chargeof['annual_inc'].apply(income_classification)
```

##### Feature on "INTEREST RATE": int_rate_classification

```python
def int_rate_classification(int_rate):
    if int_rate >= 16:
        return 'High'
    elif int_rate >= 8 and int_rate < 16:
        return 'Medium'
    elif int_rate > 0  and int_rate < 8 :
        return 'Low'
    else:
        return 'Not Known'
# Additional column is added in all 3 dataframes    
loan_past['int_rate_zone'] = loan_past['int_rate'].apply(int_rate_classification)
loan_fullypaid['int_rate_zone'] = loan_fullypaid['int_rate'].apply(int_rate_classification)    
loan_chargeof['int_rate_zone']  = loan_chargeof ['int_rate'].apply(int_rate_classification)
```

##### Feature on "LOAN AMOUNT": loan_amt_classification

```python
#---------------------------------function for loan amount classification--------------------------------------------------------------------------

def loan_amt_classification(loan_amnt):
    if loan_amnt > 23000:
        return 'Very High'
    elif loan_amnt > 15000 and loan_amnt <= 23000:
        return 'High'
    elif loan_amnt > 5000  and loan_amnt <= 15000:
        return 'Medium'
    elif loan_amnt > 0  and loan_amnt <= 5000:
        return 'Low'
    else:
        return 'Not Known'
    
loan_past['loan_amnt_zone'] = loan_past['loan_amnt'].apply(loan_amt_classification)    
loan_fullypaid['loan_amnt_zone'] = loan_fullypaid['loan_amnt'].apply(loan_amt_classification)    
loan_chargeof['loan_amnt_zone'] = loan_chargeof['loan_amnt'].apply(loan_amt_classification)
```
